```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list

list=[1,2,3,5,6,7,9,12,20,60]


def evensum(list):
    sum=0                #sum should be locally available
    for i in list:
        if i%2==0:
            sum=sum+i    # sum of even number
            #print(i)
    return sum
n=evensum(list)
print("The sum of the even number is:",n)
```

```python
# 2. Create a Python function that accepts a string and return the reverse of that string

str=input("Enter a string here:")

def string_rev(str):
    rev=""
    for i in str:
        rev=i+rev
    return rev
rev=string_rev(str)
print("The reverse string is:",rev)
    
```

```python
# 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number

def squar(n):
    str=[]
    for i in n:
        str.append(i**2)
    yield str
list(squar([5,10,15]))
 
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

numbers=list(range(1,201))
def prime(num):
    if num<=1:
        return False
    for i in range(2,int(num**0.5)+1):
        if num %2==0:
            return False
        return True
print("Prime number from 1 to 200")
for num in numbers:
    if prime(num):
        print(num, end=" ")
        
```

```python
#5. Create iterator class in Python that generates the Fibonacci sequence up to specific number of terms.

def fib(n):      #function declaration
    
    a=0
    b=1
    for i in range(n):   
        yield a
        a,b=b,a+b
    
f=fib(10)
next(f)   #displaying next value one at a time
```

```python
#7 . Implement a generator function that reads a file line by line and yield each line as as string
file=open('new.txt','r')
def read_file(file):
    for line in file:
        yield line
str=read_file(file)
next(str)
```

```python
# 8. Use lambda function in Python to sort a list of tuples based on the second element of each tuple.
std_tup=(('Arun',23),('Madhu',18),('Raju',35))
print("Initial list:",std_tup)
sort_tup=sorted(std_tup, key=lambda x: x[1])
print("Sorted list:",sort_tup)
```
